### 3.2 PNL

```python
l_gene_name = []
l_p_value_foward = []
l_p_value_backward = []
from causallearn.search.FCMBased.PNL.PNL import PNL
X = np.array(X)
for i in range(X.shape[1]):
    pnl = PNL()
    p_value_foward, p_value_backward = pnl.cause_or_effect(X[:,i].reshape(X.shape[0],1), y.reshape(X.shape[0],1))
    if p_value_foward != [0] or p_value_backward != [0]:
       # print(l_gene[i])
       # print(p_value_foward, p_value_backward)
        l_gene_name.append(l_gene[i])
        l_p_value_foward.append(p_value_foward)
        l_p_value_backward.append(p_value_backward)
df = pd.DataFrame([l_gene_name,l_p_value_foward,l_p_value_backward])
df.to_csv("p_value_breast_cancer_npl_hybird.csv")
for i in range(len(l_gene_name)):
    if l_p_value_foward[i]<0.05:
        print("forward:",l_gene[i])
    if l_p_value_backward[i]<0.05:
        print("backward:",l_gene[i])
```
